### Zadanie 2 - metoda dystrybuanty odwrotnej

```python
def odwrotna(x, lamb, alfa):
    return (lamb*(1-(1-x)**(1/alfa)))/(1-x)**(1/alfa)
```

```python
u=[random.random() for i in range(1000)]
```

```python
lamb=1
alfa=3
y_u=[odwrotna(u[i],lamb,alfa) for i in range(len(u))]
```

```python
Eu=sum(u)/len(u)
Varu=sum([(u[i]-Eu)**2 for i in range(len(u))])/len(u)
Varu
```

```python
plt.scatter(y_u,u)
```
